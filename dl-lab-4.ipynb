### AIM: write a program to implement auto encoders for both binary and real valued inputs clearly show the loss after each iteration

#### Binary inputs

Loss function: cross entropy 

−[(target)log(prediction)+(1−target)log(1−prediction)]

```python
import math
import random
from keras.models import Sequential
from keras.layers import Dense, Activation
import numpy as np
```

```python
# random input size
n = random.randint(4,20)
inp = [random.randint(0,1) for i in range(n)]

k = random.randint(1,n-1)
print("n: ", n)
print("inp: ",inp)
print("k: ", k)
```

```python
# input and output
x = np.array(inp)
y = np.array(inp) 

# Neural network
model = Sequential()
model.add(Dense(k, input_dim=n, activation='sigmoid'))
model.add(Dense(n, activation='sigmoid'))

# Compile the model and calculate its accuracy:
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) 

# Print a summary of the Keras model:
model.summary()
```

```python
x = x.reshape((1,n))
y = y.reshape((1,n))
print("x: ", x)
print("y: ", y)
```

```python
history = model.fit(x, y, epochs=10, batch_size=1)
```

```python
y_pred = model.predict(x)
print(y_pred[0])
```

```python
loss, accuracy = model.evaluate(x, y)

print("Loss on test data: ", loss)
```

#### Real inputs

Loss function: mean_squared_error

```python
# random input size
n = random.randint(4,20)
inp = [random.randint(0,100) for i in range(n)]

k = random.randint(1,n-1)
print("n: ", n)
print("inp: ",inp)
print("k: ", k)
```

```python
# input and output
x = np.array(inp)
y = np.array(inp) 

# Neural network
model = Sequential()
model.add(Dense(k, input_dim=n, activation='sigmoid'))
model.add(Dense(n, activation='sigmoid'))

# Compile the model and calculate its accuracy:
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy']) 

# Print a summary of the Keras model:
model.summary()
```

```python
x = x.reshape((1,n))
y = y.reshape((1,n))
print("x: ", x)
print("y: ", y)
```

```python
history = model.fit(x, y, epochs=10, batch_size=1)
```

```python
y_pred = model.predict(x)
print(y_pred[0])
```

```python
loss, accuracy = model.evaluate(x, y)

print("Loss on test data: ", loss)
```
