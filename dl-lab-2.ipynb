# function: X<sup>2</sub>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(start=-1, stop=1, step=0.01)
y = [i*i for i in x]
```

```python
plt.xlim(-1, 1)
plt.ylim(0, 1)
plt.grid()
plt.plot(x, y, marker="o", markersize=1, markeredgecolor="blue")
plt.show()
```

```python
import random
```

```python
def gd(x):
  epoch = 30
  ita = 0.1
  solutions, scores = [], []
  solution = x[random.randrange(0,len(x)-1)]
  for i in range(epoch):

    grad = 2*solution # 2*x
    solution = solution - ita*grad
    solution_eval = solution**2 # x^2

    solutions.append(solution)
    scores.append(solution_eval)

    print('>%d f(%s) = %.5f' % (i, solution, solution_eval))
    plt.plot(x, y, marker="o", markersize=1, markeredgecolor="blue")
    plt.plot(solutions, scores , '.-', color='red')
    plt.show()
    if(solution_eval.round(4) == 0.0000):
      print("found")
      plt.plot(x, y, marker="o", markersize=1, markeredgecolor="blue")
      plt.plot(solutions, scores , '.-', color='red')
      plt.show()
      break
  return [solutions, scores]

plt.xlim(-1, 1)
plt.ylim(0, 1)
plt.grid()
newY, score = gd(x)
```

# function: X<sup>2</sup> + Y<sup>2</sup> 

```python
x = np.arange(start=-1, stop=1, step=0.01)
y = np.arange(start=-2, stop=2, step=0.01)
z = []
for i in range(len(x)):
  z.append(x[i]**2 + y[i]**2)
```

```python
from mpl_toolkits import mplot3d

def f(x, y):
   return x ** 2 + y ** 2

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
```

```python
from matplotlib import pyplot
pyplot.contourf(X, Y, Z, levels=50, cmap='jet')
pyplot.show()
```

```python
from matplotlib import cm
from matplotlib import pyplot
import math

def f(x, y):
   return x ** 2 + y ** 2

def gd(tx, ty):
  epoch = 30
  ita = 0.1
  solutionsX,solutionsY, scores = [], [], []
  ind = random.randrange(0,len(x)-1)
  solutionX = tx[ind]
  solutionY = ty[ind]

  alpha = 0
  e = -1**7
  for i in range(epoch):

    gradX = 2*solutionX # 2*x
    gradY = 2*solutionY # 2*y
    grad = gradX + gradY
    alpha = grad*grad

    ita = ita/(math.sqrt(alpha) + e)

    solutionX = solutionX - ita*grad
    solutionY = solutionY - ita*grad
    solution_eval = solutionX**2 + solutionY**2 # x^2

    solutionsX.append(solutionX)
    solutionsY.append(solutionY)
    scores.append(solution_eval)

    print('>%d f(%s, %s) = %.5f' % (i, solutionX,solutionY, solution_eval))

    if(solution_eval.round(4) == 0.0000):
      print("found")
      break
  return [solutionsX,solutionsY,scores]

newX,newY, score = gd(x, y)

X, Y = np.meshgrid(newX, newY)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap=cm.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
```
